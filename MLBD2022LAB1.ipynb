```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing the dataset

df = pd.read_csv("/content/Bank_Modelling.csv")
x = df.iloc[:,3:13].values
y = df.iloc[:,13].values
print(x)
print(y)
```

```python
#Encoding categorial data

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

labelencoder_X_1 = LabelEncoder()
x[:,1] = labelencoder_X_1.fit_transform(x[:,1])

x
```

```python
labelencoder_X_2 = LabelEncoder()
x[:,2] = labelencoder_X_2.fit_transform(x[:,2])

x
```

```python
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([("Country", OneHotEncoder(),[1])], remainder = "passthrough")
x = ct.fit_transform(x)

x.astype(int)
x
```

```python
x = x[:, 1:]
x
```

```python
from pandas.core.common import random_state
#splitting the daatset into the training set and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state = 0)

X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

X_train
X_test
```

**Artificial Neural Network**

```python
#import the keras libraries and packages
import keras
from keras.models import Sequential #importing sequential model
from keras.layers import Dense #fully connected layer structure
```

```python
#Initialising the ANN
classifier = Sequential()#classifier is ANNA which is sequential
```

```python
#Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11)) #give a dense layer, some small number is initialized to the weight, 11 units are given to 6 hidden unit 11 units 1 initializer 11+1= 12/2
#relu rectitfies  0 and 1
```

```python
#adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))#one more hidden layer
```

```python
#Adding output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))#output will be sigmoidal
```

```python
#compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy']) #optimizer is the algo used to reduce the error here we use adam algo, error we used to calculate here is binary_crossentropy, based on accuracy 
```

```python
#fitting the ANN to the training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100 )
```

```python
#part 3 -Making prediction and evaluating the model
#predicting the test results
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_pred = (y_pred > 0.5)#ypred greater then 0.5 true else false
y_pred
```

```python
#making the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
